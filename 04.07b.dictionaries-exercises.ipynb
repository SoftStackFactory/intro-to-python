# 1. Populating a dictionary

#### Create a dictionary by using all of the given variables. 

>Hint: Some of the `keys` in the final dictionary will have `values` that require combinations of multiple varibles found below! 

>The `name` property of `my_dict` for example, relies on `first` and `last` `name`

```python
first_name = 'John'
last_name = 'Doe'
productive_hobby = 'Python'
sporty_hobby = 'gym'
age = 82
```

#### Your Solution:

```python
# I've hard coded the answer below, recreate that, using the variables above
my_dict = {
    'name': first_name + ' ' + last_name,
    'age': age,
    'hobbies': [productive_hobby, sporty_hobby]
}
```

#### Test Solution:

```python
assert my_dict == {
        'name': 'John Doe',
        'age': 82,
        'hobbies': ['Python', 'gym']
    }
```
