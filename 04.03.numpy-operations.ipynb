## Elementwise Operations

With numpy you can easily perform array with array arithmetic, or scalar with array arithmetic. 

**References:**

- [Numpy array operations](https://scipy-lectures.org/intro/numpy/operations.html)

- [Numpy methods and attributes](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.html)

Let's see some examples:

```python
print(arr + arr)
```

```python
print(arr - arr)
```

```python
# Warning on division by zero, but NOT an error!
# Just replaced with nan
print(arr / arr)
```

```python
# Also warning, but not an error, instead infinity as `np.inf`
print(1 / arr)
```

```python
print(arr ** 3)
```

### Numpy array methods

#### Sum of all elements in the numpy array
```python 
print(arr.sum())
```

#### Mean of all elements in the numpy array
```python 
print(arr.mean())
```

#### Standard deviation of all elements in the numpy array
```python 
print(arr.std())
```

#### Variance or standard deviation squared of all elements in the numpy array
This tells us the spread of the values, or in other words how far each point is from each other in general

```python 
print(arr.var())
```
