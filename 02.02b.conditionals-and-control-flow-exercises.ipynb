# 2. `if-elif-else`
Fill missing pieces (`____`) of the following code such that prints make sense.

```python
name = 'John Doe'
```

#### For the given value of `name = John Doe`, use the `len()` function to fill in each of the missing conditions, so that `length_description = semi-short`.

```python
if len(name) > 20 :
    print(name, 'is more than 20 chars long.')
    length_description = 'long'
elif len(name) > 15 :
    print(name, 'is more than 15 chars long')
    length_description = 'semi long'
elif len(name) > 10 :
    print(name, 'is more than 10 chars long')
    length_description = 'semi long'
elif len(name) > 7 and len(name) < 11:
    print(name, 'is 8, 9 or 10 chars long')
    length_description = 'semi short'
else:
    print(name, 'is a short name')
    length_description = 'short'
```

#### Test your Solution

```python
assert length_description == 'semi short'
```
