## Matplotlib

The first library we will be looking at is called `matplotlib`, we will be making plots thorugh its `pyplot` sub-module.

We can access `pyplot` by using the `.` operator on the `matplotlib` `import` statement. This gives us: 
> `import` `matplotlib.pyplot`. 

We should add the familiar `as` parameter to alias that entire library import down to just `plt`.
> `import` `matplotlib.pyplot` `as` `plt`

This simplifies invoking a `histogram` considerably, we go from typing out `matplotlib.pyplot.hist()` *every single time.* to just `plt.hist()`, much better.

### 1. Import `matplotlib.pyplot` package

```python
import matplotlib.pyplot as plt
```

<br>

### 2. Basic Matplotlib Commands

We can create a very simple `Histogram` plot using the `Age` feature of the Titanic Dataset above, stored in `df`.

[Matplotlib Hist documentation](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html)

#### Single plot on `figure` and `axis` objects.

There is a very simple analogy to remember the lines of code that you have to write to create and view a matplotlib plot in jupyter notebook. Think of making and displaying a plot, the way an artist completes a painting.

1. Stand up the Easel
2. Place Blank Canvas on the easel
3. "Draw" whatever plot on Canvas
4. "Hang" Canvas on Wall

<br>

### 3. Simple Plot of the `Age` Feature

```python
# Stand up the Easel
fig = plt.figure()

# Place Blank Canvas on the easel
ax = fig.gca()

# "Draw" a histogram onto the Canvas
ax.hist(df['Age'])

# Hang Canvas on Wall
plt.show()
```

<br>

### 4. With Labels and Titles

Things get a whole lot more complex looking, but just because the ode below `looks` harder, doesn't make it so. I would not recommend remembering the specific commands to create labels and add titles, especially with the `seaborn` library that we will see in month 2.

```python
# Set a figure
fig = plt.figure()

# Grab `ax` object
ax = fig.gca()

# Fig Title
fig.suptitle("Histogram of Ages")

# Ax Title
ax.set_title("For all Passengers Aboard RMS Titanic")

# Plot data
ax.hist(df['Age'])

# Labels
ax.set_xlabel("Passenger Age")
ax.set_ylabel("Frequency")


# Clean up output
plt.show()
```

### 5. Saving figures
Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF. 

#### To save a figure to a file we can use the `savefig` method on the `fig` object we created above:

```python
filePath = "./data/age-histogram.png"
fig.savefig(filePath)
```

#### Here we can also optionally specify the DPI and choose between different output formats:

```python
filePath = "./data/age-histogram-200-dpi.png"
fig.savefig(filePath, dpi=200)
```

## Seaborn

The second library we will be looking at is called `seaborn`.

We should add the familiar `as` parameter to alias that entire library down to just `sns`.
> `import` `seaborn` `as` `sns`

### 1. Simpler Plot of the `Age` Feature Using Seaborn

Seaborn's histogram plot function is called `distplot`.

[Seaborn distplot documentation](https://seaborn.pydata.org/generated/seaborn.distplot.html)

We have to chain .dropna() so seaborn doesn't try to plot Nan values — other wise the plot will break.

```python
# "Draw" a histogram, seaborn handles ax and fig
sns.distplot(df['Age'].dropna())

# Remove unwanted output
plt.show()
```

### 2. Combine approaches for added control.

```python
# Stand up the Easel
fig = plt.figure()

# Place Blank Canvas on the easel
ax = fig.gca()

# Add labels and title
ax.set_title('A Distribution Plot for the Age Variable')
ax.set_ylabel('Frequency')

# "Draw" a histogram onto the Canvas, pass in ax to ax seaborn parameter
sns.distplot(df['Age'].dropna(), ax=ax) # don't forget to drop NAN values

# Hang Canvas on Wall
plt.show()
```

## Pandas Built-in

The third library we will be looking at you're already familiar with, but its more versatile than you previously thought!

### 1. Even Simpler Plot of the `Age` Feature

We have to chain `.plot.hist()` on whatever column we are interested in viewing a `histogram` for.

```python
# "Draw" a histogram, seaborn handles ax and fig
df['Age'].plot.hist()

# Remove unwanted output
plt.show()
```

### 2. Combine approaches for added control.

```python
# Stand up the Easel
fig = plt.figure()

# Place Blank Canvas on the easel
ax = fig.gca()

# Add labels and title
ax.set_title('A Distribution Plot for the Age Variable')
ax.set_ylabel('Frequency')

# "Draw" a histogram onto the Canvas, pass in ax to pandas ax parameter
df['Age'].plot.hist(ax=ax)

# Hang Canvas on Wall
plt.show()
```
