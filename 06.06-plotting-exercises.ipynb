### Matplotlib

#### Set-up
We will be discussing many different kinds of plots in month 2 and the `histogram` is among the most important. A `histogram` describes the `distribution` of values for a `continuous` variable.

> A `pandas` column is considered to be a `continuous` variable when the values contained could take on any value along the number-line between its `min` and `max` values.

> There are many other types of data in the world other than continuous, for ex 
- `binary` labels 
  - `Male`/`Female`, 
  - `survived`/`not_survived`
- `categorical` labels 
  - `Pclass` `1`/`2`/`3`
  - `Embarked` port `C`/`Q`/`S`


#### Steps:
1. Create a `figure` object; call the `plt.figure()` function, save its output to a variable named `fig`.

2. Create an `axes` object; call the `.gca()` method on the `fig` object —`fig.gca()`, save its output to a variable named `ax`.

3. Create a `plot`; call the `.hist()` method on the `ax` object, pass in the `Fare` column `df['Fare']`, as the argument to `ax.hist('here')`, no need to save to a variable.

5. Add a label to the horizonatal `X-Axis`;  call ax.set_xlabel() , pass in "Fare"

> `axes` is the plural of `axis`; when we make an `axes` object in step `2`, we are simply creating the the horizontal `x-axis` and vertical `y-axis` that some 2 dimensions of our data will "hook" onto.

6. Call `plt.show()` to "print" the plot

> `plt.show()` captures any lingering output instead of letting it output to screen. These `out` statements you'll notice, if you run a cell with the function being called but before you save the output to a variable.

### Seaborn

#### The simplicity of Seaborn!
- `Seaborn` comes installed with the `conda` and `google colab` distriution, the following will walk you through the steps to make a simple `seaborn` `plot` that is equivalent to the one from the previous exercise.

> `seaborn` calls a `histogram`, with its own `.distplot()` function —they're very similar, in terms of function but the distplot is certainly much more detailed and soesn't pre bin the data into columns.


#### Set-up

- A `seaborn` plot will automatically derive the `x` and `y` labels of a plot based on the `pandas` `columns` that you pass in, this means less work for you! Also, some plots can consider more than 1 column of data through optional parameters!


- A `distplot` only requires data along the `x` axis, as the `y` data will always be derived from the `density` or `frequency` of each `x` value, as it appears in the `distribution`. 

> When you see a hump in the `y` direction, this means that the range of `x` values underneath said hump, are more common in the distribution.


#### Steps
Make a `seaborn` `.distplot()` on `Fare` column, equivalent to `ax.hist()`

1. `import` `seaborn` `as` `sns`
2. pass `df['Fare']` into `sns.distplot()`
3. Add `plt.show()` underneath to `print` the plot
4. Yup, `3` lines of code. Welcome to `Seaborn`!

### Pandas built-in

#### The further simplicity of `pandas`!
- `pandas` has some powerful tools that are great for quick visualization.

> `pandas` calls by chaining some commands together. We'll explore that below.


#### Set-up

- `pandas` has plots that work both on entire `dataframes` and single `columns`, it is very important that you know the difference as you will chain specific commands to reach the plot that you want and the data you pass-in before the first `.`, must be compatible with the plot in question. 


- `pandas` histogram can be accessed like so, notice the first `.`:

> `pd['column_name'].plot.hist()`
> or `pd['column_name'].plot(kind='hist')`


#### Steps
Make a `pandas` `.plot.hist()` on `Fare` column, equivalent to `sns.distplot()`**

1. pass `df['Fare']` by chaining `.plot.hist()`
2. Add `plt.show()` underneath to `print` the plot
3. Yup, `2` lines of code.

#### Wrap-up

- Compare the `pandas` `hist` to `sns` `distplot`. 

- The `seaborn` doensn't bin your values, but when you're looking to make a plot for a quick glimpse, a `hist` will do just fine.

- Notice that all of these plots look very similar, this is because they all "sit" ontop of `matplotlib`, with `matplotlib` being the most customizable, `seaborn` being the most sophisticated, and `pandas` being the most effective.
