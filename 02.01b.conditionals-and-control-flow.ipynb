### `and`

```python

print('True and False:', True and False)
```

````python
print('True and True:', True and True)
print('True and False:', True and False)
````

### `or`

```python
print('True or True:', True or True)
print('True or False:', True or False)
print('Fasle or False:', False or False)
```

````python
print('True or True:', True or True)
print('True or False:', True or False)
print('False or False:', False or False)
````

### `not`

```python

print('Python is cool:', not python_is_cool)
```

````python
python_is_cool = False
print('Python is cool:', not python_is_cool)
````

#### We can combine them! 
>Be sure to use parentheses to group operations!

```python

print("(True and False) and (True or False):", eval_false and eval_true)
```

````python
eval_false = True and False
eval_true = True or False
print("(True and False) or (True or False):", eval_false or eval_true)
print("(True and False) and (True or False):", eval_false and eval_true)
````

## 4. If Statements

<img width = "500px" src="https://data36.com/wp-content/uploads/2018/01/1-Python-if-statement-logic-768x374.png">

#### Basic Example

```python
# Run twice, 2nd time: change to statement = False
statement = True

if statement:
    print("statement is True")
    
if not statement:
    print("statement is NOT True")
```

````python
statement = True

if statement:
    print("statement is True")

if not statement:
    print("statement is NOT True")
````

<br>

#### `if()` statements can handle *multiple* conditions with `and` and `or`

```python
val = 3

# Take note of the parantheses!
if ((val > 0) and (val < 1)) or (val == 3):
    print('Value is positive -and- less than one --or-- value is three')
```

````python
val = 3

# Take note of the parantheses!
if ((val > 0) and (val < 1)) or (val == 3):
    print('Value is positive -and- less than one --or-- value is three')
````

### `if-elif-else`

```python
# run code with diff values for age
age = 19

if age >= 21:
    print("Person is old enough to vote AND purchase alcohol")

elif age >= 18:
    print("Person is old enough to vote BUT NOT purchase alcohol.")

else:
    print("Person is NOT old enough to vote OR purchase alcohol.")
```

````python
age = 22

if age >= 21:
    print("Person is old enough to vote AND purchase alcohol")

else: 
    if age >= 18:
        print("Person is old enough to vote BUT NOT purchase alcohol.")

    else:
        print("Person is NOT old enough to vote OR purchase alcohol.")
````

<br>

#### You can have as many `elif` statements as you need. In addition, `else` at the end is not mandatory.

```python
# Hint: Change value of language

language = 'Swedish'

if language == 'Swedish':
    special_greeting = 'Hejsan!'
elif language == 'Finnish':
    special_greeting = 'Latua perkele!'
elif language == 'Spanish':
    special_greeting = 'Hola!'
elif language == 'German':
    special_greeting = 'Guten Tag!'
else:
    special_greeting = 'Hello!'

greeting = special_greeting + ' fellow Pythonista!'
print(greeting)
```

````python
language = 'German'

if language == 'Swedish':
    special_greeting = 'Hejsan!'
elif language == 'Finnish':
    special_greeting = 'Latua perkele!'
elif language == 'Spanish':
    special_greeting = 'Hola!'
elif language == 'German':
    special_greeting = 'Guten Tag!'
else:
    special_greeting = 'Hello!'

greeting = special_greeting + ' fellow Pythonista!'
print(greeting)
````
