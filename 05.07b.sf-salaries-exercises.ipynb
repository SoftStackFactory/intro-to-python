## Bonus

**<span style="color:red">Hard: </span> How many people have the word `Chief` in their JobTitle?**
**Hint:** [`.str.contains()`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.str.contains.html)

```python
sal["Indexes"] = sal["JobTitle"].str.find("Chief")
len(sal[sal["Indexes"] != -1])

chiefs_sum = 0
for title in sal["JobTitle"]:
    if 'Chief' in title.split(" "):
        chiefs_sum += 1 
print(chiefs_sum)

sal[sal["JobTitle"].str.contains("Chief")].shape[0]
```
