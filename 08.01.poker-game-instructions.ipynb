# 5 Card Draw Poker

Let's build a simple Poker game application for fun, we are going to tackle traditional 5 card draw.


![image](https://cloud.githubusercontent.com/assets/3248824/19602814/950fcbd2-97a6-11e6-9e9d-faabeac2307f.png)


### The game

The basic game of 5 card draw is played with one deck of 52 cards, with at most 6 players, any more and you will run out of cards.

Betting typically consist of an ante, where an amount to is placed by each player at the start of the game.

Then the cards are dealt to each player from left to right one at a time until all five cards are dealt out.  

Each player reviews their hand and they decide how many cards they may want to replace.  Each player discards the cards and they are dealt replacement cards in series, then the next player discards and receives there cards, not the previous persons discarded cards.

Players may then bet again to stay in the game players must meet the new bets value.  Players may fold and leave the game if they choose not to continue to bet.

The players then evaluate their hands to see which hand has one.

More info about this type of poker version: https://en.wikipedia.org/wiki/Five-card_draw

        - List

- **Services**:
