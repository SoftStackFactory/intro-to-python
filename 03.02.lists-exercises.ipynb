# 1. Indexing and Slicing

````python
years = [1998,1999,2000,2001,2002,2003]
````

<br>

**Use `list` `indexing` Grab the first value in `a_list`**

````python
first_year = years[0]
````

<br>
<br>

**Use `list` `slicing` to grab the last three values in `a_list`**

````python
max_idx_of_years = len(years) - 1
last_three_years = years[max_idx_of_years - 2 : max_idx_of_years + 1]

# OR 
len_of_years = len(years)
last_three_years = years[len_of_years - 3 : len_of_years]

print(first_year, last_three_years)
````

<br>
<br>

**This next cell will make sure that the variables `first_year` and `last_three_years` have been populated with the appropriate values**

> Hint: You can always use `print()` statements to check what is stored inside of a variable, if the `assert` does not pass!

````python
assert first_year == 1998
assert last_three_years == [2001,2002,2003]
````

# 3. Create a new list without modifiying the original one


````python
original = ['I', 'am', 'learning', 'hacking', 'in']
````

<br>

### Steps Broken Down


Copy original List by value not by reference 


````python
modified = original.copy()
````

```python
Update 'hacking' to 'lists'
```

````python
modified[3] = 'lists'
````

```python 
Append 'Python'
```

````python
modified.append('Python')
````

<br>

#### Test Your Solution

````python
assert original == ['I', 'am', 'learning', 'hacking', 'in']
assert modified == ['I', 'am', 'learning', 'lists', 'in', 'Python']
````
